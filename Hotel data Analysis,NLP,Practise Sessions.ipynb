```python
#Write a function that will give sum of numbers in the list
#without using any predefined python function
def sum_of_list(numbers):
  sum=0
  for i in numbers:
    sum=sum+i
  return sum
print(sum_of_list([3,5,7]))
```

```python
#Write the function that will give count of vowels present in the given string/text
#Not to use predefined python libraries/functions
def count_vowels(text):
  vowels="aeiou"
  count=0
  for x in text:
    if x in vowels:
      count+=1
  return count
count_vowels("Hello,how r u?")
```
